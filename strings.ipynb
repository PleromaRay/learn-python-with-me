```python
# some python string methods
#create some example strings
e_str1 = "Hello"
e_str2 = "WORLD"
e_str3 = ", again"
e_str4 = "1233"
e_str5 = "32"
e_str6 = "hello world, again"
e_str7 = "    hello    "
e_str8 = "    world,   "
```

```python
#test strings
print(e_str4.isalpha()) #true or false?
print(e_str4.isdecimal()) #true or false?
print(e_str4.isnumeric()) #true or false?
print(e_str5.isnumeric()) #keep float and integer separate
```

```python
print(e_str2.isupper())
print(e_str1.isupper())
print(e_str3.islower())
```

```python
#searching for contents inside strings
#return integer values describing position of finding and -1 if no finding occur
print(e_str6)
print(e_str6.find("world"))
print(e_str6.rfind("lo"))
print(e_str6.find("aircraft"))
```

```python
#count the occurrence of a string inside a string
print(e_str6.count("l"))
```

```python
#changing or replacing the content of a string, creating a new copy
#of the transformed string
print(e_str1.capitalize())
print(e_str2.lower())
print(e_str3.upper())
print(e_str6.replace("l", "1"))
```

```python
#method that can split string into a list of strings
e_str10 = e_str7 + e_str8
print(e_str10)
list_str = e_str10.split()
print(list_str)
for str in list_str:
    print(str)
```

```python
#methods to remove whitespace
print(e_str10) #would print with whitespaces
print(e_str7.lstrip() + e_str8.rstrip())
print(e_str7.strip() + " " + e_str8.strip())
print(e_str10.strip())
```

```python
#libraries import datetime module as dt
import datetime as dt

# The date object
# Define date data through a function or as data insertion 
# In this case we would use integers created from string time
# For example older standard 1996-10-27, or 27/10/1996 or I-96/27-10
current_date = dt.date.today()
birth_date = dt.date (1994, 12, 27)

#data handling involves unifying many date and time standards
print(birth_date.month)
print(current_date.day)
print(birth_date.year)
print(birth_date)
print(current_date)
```

```python
#let's take a look at a time object
obj_t = dt.time() 
print(obj_t)
print(dt.datetime.now())
# .now(tzinfo=None)
```

```python
#create a date obj with today's date in your local time
local_today = dt.date.today()
print (local_today)
```

```python
#create a date time object with today's date and zero time
local_today = dt.datetime(local_today.year, local_today.month, local_today.day)
print(local_today)
```

```python
current_datetime = dt.datetime.now()
print(current_datetime)
```

```python
#use datetime to calculate the time today and now
time_now = dt.datetime.now()
time_since_obj_t = time_now - local_today
print(time_since_obj_t)
```

```python
#remove microseconds
time_since_obj_t -= dt.timedelta(microseconds = time_since_obj_t.microseconds)
print(time_since_obj_t)
```

```python
#cal time differences using datetime objects
days_old = current_date - birth_date
days_old = days_old.days
print(days_old)
```

```python
#cal time constructs such as years
years_old = days_old //365 #it is a python integer
print(years_old)
```

```python
# Use the strftime method to edit output from Datatime objects which returns a string representing the datetime, controlled by an 
# explicit format for the string. In the manual there is a large number of codes for different output and formatting.
# Reuse time_now datetime with current date and time
print(time_now)

# date
ymd = time_now.strftime("%y/%m/%d")
print("date:", ymd)
```

```python
# time
time = time_now.strftime("%H:%M:%S")
print ("time:", time)
```

```python
# Datetime
datentime = time_now.strftime("%m/%d/%Y, %H:%M:%S")
print ("date and time:", datentime)
```

```python
# this is one way to format string output from a datatime object

# Let's take a look a the corresponding .strptime() method to format input to a datetimeobject from a string

# Create a string with an obscure date and time format

str_date_time = "06, June, 1999 12:45" # 
print ("string ", str_date_time)
print ("type" , type(str_date_time))
```

```python
# Use .strptime() to create a datetime object from the string. strptime needs an existing datetime object.

dt_object = time_now.strptime(str_date_time, "%d, %B, %Y %H:%M")
print("datetime object ", dt_object)
print("type", type(dt_object))
```
