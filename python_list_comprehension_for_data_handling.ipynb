```python
# List comprehension is a different way to express a code that will yield the same result
# as an ordinary list with a for-loop. Helps to make your code more readable, easier, and 
# better depending on the task

#List comprehensions vs. For-loops

#create a list
a_list = [4,5,6,7,8]
print(a_list)
```

```python
#show the for-loop 3 lines of cod vs list comprehension one line f code
b_list = []
for element in a_list:
    b_list.append(element + 1)
print (b_list)
```

```python
#list comprehension
a_list_plus1 = [element + 1 for element in a_list]
print(a_list_plus1)
```

```python
#this is optimal setting for making use of list comprehension
#it removes about 50% of the code and for small list with fits in Ram
#list comprehension can increase code speed by about 100%

#list comprehension for multiple dimensions may quickly grow in complexity and may
#cause issues with readability 
```

```python
#2D list comprehension by creating a list of lists
list_of_lists = [[2, 3, 4], [5, 6, 7], [8, 9, 10]]
print(list_of_lists)
```

```python
list_of_lists_minus1 = []
for list in list_of_lists:
    c_list = []
    for element in list:
        c_list.append(element-1)
    list_of_lists_minus1.append(c_list)
print(list_of_lists_minus1)
```

```python
#let's show the proper way to type using list comprehension

list_of_lists2 = [[element+1 for element in list] for list in list_of_lists]
print(list_of_lists2)
```

```python
#2D is considered the generally highest dimension where list dimension remain pythonic
#even for non-trivial examples
```

```python
#3d list (for-loops)

list_3D = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(list_3D)
```

```python
list_3D_minus1 = []
for list in list_3D:
    d_list = []
    for list in list:
        e_list = []
        for element in list:
            e_list.append(element-1)
        d_list.append(e_list)
    list_3D_minus1.append(d_list)
print(list_3D_minus1)
```

```python
#list comprehension
list_3D_plus1 = [[[element+1 for element in list] for list in list] for list in list_3D]
print(list_3D_plus1)
```

```python
#in this case, the 3D list comprehension is good but can get really complex fast.
#not following the strict naming scheme may also result in readability issues for more complex expressions
#general recommendation is to stay in 2D because 99% of coders should understand 2D code.
```
