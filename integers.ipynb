```python
# python integer
#create a python integer

whole_num = 5
print (whole_num)
```

```python
# create more integers and add them together to create more integers
num_two = 20
num_three = whole_num + num_two
print (num_three)
```

```python
#subtraction
num_four = num_three - whole_num
print (num_four)
```

```python
# multiplication
num_five = num_four * whole_num
print(num_five)
```

```python
#division
num_six = num_five / whole_num
print(num_six) # answer is a float because it has a decimal 
```

```python
#type checking
print(type(num_six))
print(type(num_four))
```

```python
#floor division #removes decimal and returns an integer instead of a float
num_seven = num_five//whole_num
print(num_seven)
print(type(num_seven))
```

```python
# integers can be used to keep track of python loops
for x in range (1, 6):
    print (x)
```

```python
#other data types can be converted as integers if the data type is suitable
num_eight = int(5.5) #cost a float 5.5 as an integer with the value 5
num_nine = int("3") #cost the string as integer with the value 3
print (num_eight)
print (type(num_eight))
print(num_nine)
print(type(num_nine))
```
