```python
#python floats
#create a python float with assignmnet
real_num = 5.3
print(real_num)
```

```python
#create more floats and combine them to create more float through assignmnet
fp_two = 3.2
print(fp_two)
```

```python
#addition
fp_three = fp_two + real_num
print(fp_three)
```

```python
#substration
fp_four = real_num - fp_two
print(fp_four)
```

```python
print(round(fp_four, 1))
```

```python
#multiplication
fp_four_b = round(fp_four, 1)
fp_five = fp_four * fp_two
fp_five_b = fp_four_b * fp_two
print(fp_five)
print(fp_five_b)
```

```python
round(fp_five, 1)
round(fp_five_b, 1)
```

```python
fp_six = fp_three/fp_two
print(fp_six)
```

```python
res = 1.0
res_b = 1.0
for x in range (1, 6):
    res = res * x * fp_four
    res_b = res_b * x * fp_four_b
    print(x * fp_four)
    print(x * fp_four_b)
    print (res)
    print(res_b)
    print (res - res_b)
```

```python
#increase float size using loops
res = 1.0
res_b = 1.0
for x in range (1, 66):
    res = res * x * fp_four
    res_b = res_b * x * fp_four_b
    print(x * fp_four)
    print(x * fp_four_b)
    print (res)
    print(res_b)
    print (res - res_b)
```

```python
#increase float size even more
res = 1.0
res_b = 1.0
for x in range (1, 666):
    res = res * x * fp_four
    res_b = res_b * x * fp_four_b
    print(x * fp_four)
    print(x * fp_four_b)
    print (res)
    print(res_b)
    print (res - res_b)
```

```python
#conclusion
#if floating point unit cannot properly define a number, rounding that number will not change
fp_seven = float (5)
fp_eight = float ("3.3")
print(fp_seven)
print(type(fp_seven))
print(fp_eight)
print(type(fp_eight))
```

```python
#convert suitable python float to python tuple with two python integers
fp_nine = 0.5
ratio_tuple = fp_nine.as_integer_ratio()
print(ratio_tuple)
print(str(ratio_tuple[0]) + "/" + str(ratio_tuple[1]))
```
