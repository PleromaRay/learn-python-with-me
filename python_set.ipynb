```python
# Python set

#create a set named fruits with named fruit objects (strings) inside

fruits = {"banana", "apple", "orange", "apple", "peach", "banana"}
print(fruits)
print(type(fruits)) # duplicates will be automatically removed
```

```python
# The objects appear unordered... but are they random?

fruits_2 = {"banana", "apple", "orange", "apple", "peach", "banana"}
print(fruits_2)
print(type(fruits_2))
```

```python
# The Python set is unordered but not "random" - there is an order to the Set "under the hood"...

#add more fruits with the .add() function

fruits.add("lemon")
fruits.add("banana")
print(fruits)
```

```python
# add multiple objects to a set with the update function

fruits.update(["pear", "citrus", "peach"])
print (fruits)
```

```python
#copying a set
#you can do this using the .copy() function
fruitage = fruits.copy()
print(fruitage)
```

```python
#remove object from a set using the .remove() function

fruitage.remove("citrus")
print(fruitage)
print("test fruits")
print(fruits) # A shallow copy is sufficient for this use... the alternative is .deepcopy()
```

```python
#clear a set using the .clear() function
fruits.clear()
print(fruits) #fruit set retuns as empty
```

```python
#delete a set with the del function

del fruits
print(fruits) # returns an error as the fruit set has previously been cleared
```

```python
#loop through a set using the for loop and print the contents

for fruits in fruitage:
    print(fruits)
```

```python
#create an empty basket set and put in some fruits using math operations

mover_set = {"apple", "peach", "banana"}
print(fruitage)
basket = mover_set.intersection(fruitage)
fruitage.difference_update(basket)
print(basket)
print(fruitage)
```
