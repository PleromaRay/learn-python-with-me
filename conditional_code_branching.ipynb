```python
#create a list of integers and modify the integers
list_1 = [2,2,4,4,6,6,8,8,10,10,12,12]
for item in range(len(list_1)): list_1 [item] = list_1 [item] * 1000

print (list_1)
```

```python
#using conditional branching for a standard data handling task
list_2 = [] #results
for item in list_1:
    if item>7000:
        list_2.append("High")
    elif (item<=7000) and (item>=5000):
        list_2.append("Medium")
    else:
        list_2.append("Low")
        
```

```python
#print the result as a table using for loop
for item in range(len(list_2)):print(str(list_1[item]) + "\t" + str(list_2[item]))
```

```python
#remake the same list with the range(len) for loop
list_3 = []
for item in range(len(list_1)):
    if list_1[item]>7000:
        list_3.append("High")
    elif ((list_1[item]<=7000) and (list_1[item]>=5000)):
        list_3.append("Medium")
    else:
        list_3.append("Low")
#print result into a table using another for-loop

for item in range(len(list_3)):print(str(list_1[item]) + "\t" + str(list_3[item]))
```

```python
#stepwise conditional code branching
list_2 = []
for item in list_1:
    if item>7000: list_2.append("High")
    if (item<=7000) and (item>=5000): list_2.append("Medium")
    if item>5000: list_2.append ("Low")
print(list_2)
```

```python
#range(len) for loop version
list_2 = []
for item in range(len(list_1)):
    if list_1 [item] > 7000: list_2.append("High")
    if ((list_1 [item] <= 7000) and (list_1[item] >= 5000)): list_2.append("Medium")
    if list_1[item] < 5000: list_2.append("Low")
print (list_2)
```

```python
#descriptive statistics using conditional branching and code operators
list_u_cat = [] #list of unique categories
nop_counter = 0 #no operation counter
for item in list_2:
    if (item not in list_u_cat):
        list_u_cat.append(item)
        print(item +  " number in category :" + str(list_2.count(item)))
    else:
        nop_counter = nop_counter + 1
print("The number of unique category are: ", len(list_u_cat))
print("The number of no-operation passes are: ", str(nop_counter))
print("The unique categories are ")
print(*list_u_cat)
```

```python
#dictionary version
desc_statistics = {}
for item in list_2:
    if item in desc_statistics:
        desc_statistics [item] += 1
    else:
        desc_statistics [item] = 1
#print a sorted output

for item in sorted(desc_statistics, key = desc_statistics.get):
    print((desc_statistics[item], item))
    
```

```python
#match-case pattern
list_4 = []
#string to numbers matching
for string in list_2:
    match string:
        case "High": list_4.append(3)
        case "Medium": list_4.append(2)
        case "Low": list_4.append (1)
        case _: list_4.append(0)

print(list_4)
```

```python
#add logic to the match-case pattern
list_2 = []
for item in list_1:
    match item:
        case item if item >7000: list_2.append("High")
        case item if item <=7000 and item >=5000: list_2.append("Medium")
        case item if item <5000: list_2.append("Low")
        case _ :list_2.append("Unknown")
print(list_2)
```
