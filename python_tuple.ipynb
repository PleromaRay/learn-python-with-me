```python
# Create some tuples

Tuple1 = (12345)  # returns error if a tuple is expected
print(Tuple1)
print(type(Tuple1))
```

```python
Tuple1 = (12345, 1)
print(Tuple1)
print(type(Tuple1))
```

```python
# Create more Tuples

Tuple2 = ("hello", "world")
Tuple3 = ("hello", 12345, "world") # Tuples can hold arbitrary python objects
Tuple4 = (1, 4, 2.73, 3.1) # Create a tuple with some floats and integers
print(Tuple2 + Tuple3)
print(Tuple1 + Tuple4)
```

```python
#create nested tuples i.e tuples within tuples
Float1 = 3.91
Integer1 = 9
Tuple5 = (Float1, Integer1, ("hello", "world"))
print(Tuple5)
```

```python
print(Tuple5[0])
print(Tuple5[1])
print(Tuple5[2])
print(Tuple5[2][1]) #access the second element of the nested 
```

```python
# Access the contents of a tuple by referring to the order of the items, indexed from 0 to the number of items minus 1. 

print(Tuple3[0])
print(Tuple3[2])
```

```python
# len() function for tuples
print(len(Tuple2))
print(len(Tuple4)) #returns an integer #number of items inside the tuple
```

```python
# Tuples can be accessed as “ranges” using the operator :

print(Tuple4[0:2])
```

```python
#loop tuples

for items in Tuple4:
    print(items)
```

```python
# Use the index positions of a tuple, and use the len() function and loops.
for pos in range(len(Tuple3)):
    print(str(pos) + " " + str(Tuple3[pos])) 
```

```python
#tuple unpacking, which is about making variables from a Tuple - something often necessary considering returns
# from libraries.
Float2, Integer2, Tuple6 = Tuple5
print(str(Float2) + " " + str(Integer2) + " " + str(Tuple6))
```

```python
str_1, str_2 = Tuple6
print(str_1 + " " + str_2)
```

```python
# show a custom statistics function with multiple returns

def desc_stats(data): #statistics library (desc_data)
    from statistics import mean, median, stdev
    average = mean(data)
    med = median(data)
    st_dev = stdev(data)

    return average, med, st_dev

print (desc_stats(Tuple4))
```

```python
#unpack the tuple return

mean, med, st_dev = desc_stats(Tuple4)
print(str(mean) + " " + str(med) + " " + str(st_dev))
```
