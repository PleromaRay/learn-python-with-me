```python
#python while-loop

# Create a first While loop
Loopcounter = 0   # Integer variable to keep track on the number of repetitions or iterations of the while loop.
while Loopcounter < 10:
    Loopcounter = Loopcounter + 1
    print(Loopcounter)
    
print("This is the final value of the Loopcounter integer variable" + ": " + str(Loopcounter))
```

```python
# While loops, infinite while loops and break statementsâ€¦
# While a while-loop is a great tool for keeping a program running indefinitely or persistently, a "forever loops" may also lock a 
# program in execution mode. One way to handle this is to insert break statements.

# This while loop creates a possible "infinity-loop" without the "break-exit" from the loop
Loopcounter2 = 10
while Loopcounter2 > 5:
    Loopcounter2 = Loopcounter2 + 1
    print(Loopcounter2)
    if Loopcounter2 == 20:
         break

print("This is the final value of the Loopcounter2 integer variable" + ": " + str(Loopcounter2))
```

```python
# "algorithmic" way to use While-loops with Python lists
list_tested = []
list_untested = ["cat", "dog", "horse", "bull", "cow"]
while len(list_untested)>0:
	print("testing a: " + str(list_untested[0]))
	list_tested.append(list_untested.pop(0))

print("all these animals have been tested:")
print(*list_tested, sep =", ") #sep means separated
```

```python
# A similar design can be implemented with Python sets
set_original = {"cat", "dog", "horse", "bull", "cow"}
set_tested = set()
set_untested = set_original.difference(set_tested)

print("Original set: " + str(set_original))
```

```python
while len(set_untested)>0:
    set_tested_memory = set_tested.copy() # Creates a set memory between the loop iterations
    set_tested.add(set_untested.pop())  # Moves a random "animal" from the tested set to the untested set
    print("Tested animal this iteration: " + str(set_tested.difference(set_tested_memory)))
    print("Untested animals:" + str(set_untested))
    print("Tested animals: " + str(set_original.difference(set_untested)))
```
