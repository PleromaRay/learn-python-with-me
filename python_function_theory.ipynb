```python
#simple pythonic function
#make a list with numeric values
list_3 = [2, 2, 4, 4, 6, 6, 8, 8]
```

```python
# Make the function
def Calcmean(list = []):
    Sum = 0.0 # a float
    for item in range(len(list)): Sum = Sum + list[item]
    Ourmean=Sum/len(list)
    return Ourmean

# Use the function
Ourmean = Calcmean(list_3)
print(Ourmean)
```

```python
#as an alternative you can use
print(Calcmean(list_3))
```

```python
#to shorten the function
def Calcmean_2(list=[]):
    return sum(list)/len(list)

print(Calcmean_2(list_3))
```

```python
#Advanced functions
#count the number of unique items in a one-dimensional list
def count_unique_items(collection):
  desc_stats = {}  # Dictionary and the return item
  for item in collection:
        if item in desc_stats:
            desc_stats[item] +=1
        else:
            desc_stats[item] = 1
  return desc_stats
    
print(count_unique_items(list_3))
```

```python
desc_statistics2 = count_unique_items(list_3)
for word in sorted(desc_statistics2, key=desc_statistics2.get):
    print((desc_statistics2[word], word))
```

```python
#special "fixed" 2D uneven multi-type list of lists
list_special = [[1, 3, 5, "nine"], [2.32, 3, 4, "nine", "nine", "six"], [1, 4, 4, 3, "six"]]
```

```python
# Creating a standard function for adding dictionaries "together".
def add(dictionaryA, dictionaryB):
    dict = {key: dictionaryA.get(key, 0) + dictionaryB.get(key, 0) for key in set(dictionaryA) | set(dictionaryB)}
    return dict
```

```python
# function that create dictionaries and add them together
def count_unique_items2(collection):
    dict_unique_items = {}
    for items in collection:
        dictitem = count_unique_items(items)
        dict_unique_items = add(dict_unique_items, dictitem)
    return dict_unique_items

descript_stats = count_unique_items2(list_special)

for word in sorted(descript_stats, key= descript_stats.get):
    print((descript_stats[word], word))
```

```python
#Test positional arguments vs. keyword arguments
#Positional arguments are common in older algorithmic code
#keywords arguments creates safety, certainty, and security in code execution

# Test positional arguments with a simple algorithm.
def test_args (alfa, bravo):
    caesar = alfa - bravo
    return caesar

print(test_args(5, 3))
print(test_args(3, 9))
```

```python
# Test keyword arguments using the same algorithm and function

print(test_args(alfa=10, bravo = 20))
print(test_args(bravo=20, alfa=10))
```

```python
# Let's test multiple return objects
def test_ret(alfa, bravo):
    caesar = alfa + bravo
    delta = bravo - alfa
    echo = alfa * bravo
    return caesar, delta, echo
tuple_1 = test_ret(5, 3)
charlie, david, eric = test_ret(5, 3)
print(tuple_1)
print(str(charlie) + " " + str(david) + " " + str(eric))
```

```python
#test some *args and **kwargs
#* and ** are unpacking instructions
#* unpacks lists or collections, ** unpacks dictionaries
# * results in a tuple, and 
#** cannot be unpacked into a single variable, may be used in more complex setting
#input to *args is a list, and the output is a tuple.

#*args test
def test_args(*args):

    res_1 = 0
    res_2 = 0
    res_3 = 1
    for i in args:
        res_1 = res_1 + i
        res_2 = res_2 - i
        res_3 = res_3 * i
    
    return res_1, res_2, res_3

print(test_args(2, -1, 2))
```
