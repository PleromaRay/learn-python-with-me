```python
# Python List can contain any kind of object irrespective of its type

# Create my first Python list
list_1 = [7, 4, 3, 9, "bananas"]
print(type(list_1))
print(list_1)
```

```python
# Shows how to address list objects using the objects position inside the list
print(str(list_1[3]) + str(list_1[2]) + " " + str(list_1[4]))  # str(list[index]) instructs Python to print the content at position index "as a string"
# This is one way to control the output from a list
```

```python
# Operate on list objects using the list element notation
print(type(list_1[3]))
print(type(str(list_1[3])))
```

```python
# A List object's content may be changed by assignment:
list_1[2] = 2
print(list_1) # This prints 7, 4, 2, 9, 'bananas' as 3 is replaced by 2
```

```python
# New objects can be added to a list with the append() method
list_1.append("apples")
print(list_1)
```

```python
# Remove objects from a list using the del statement or the remove() method
del list_1[4]
print(list_1)
```

```python
list_1.remove("apples")
print(list_1)
```

```python
# Python lists can be added together using inter alia the + operator or the Extend() or Append() methods:
# Create a list_2 and list_3
list_2 = [1, 4, 5, 6, 8]
list_3 = list_1 + list_2
print(list_3)   # 7, 4, 2, 9, 1, 4, 5, 6, 8
```

```python
list_1.append(list_3)
print(list_1)
```

```python
list_3.extend(list_2)
print(list_3)  # 7, 4, 2, 9, 1, 4, 5, 6, 8, 1, 4, 5, 6, 8
```

```python
print(len(list_3))  # 14
print(len(list_1)) # 5
```

```python
# Sort list_3
print(list_3)
list_3.sort()
print(list_3)
# The list has been sorted based on the value of the integers in the list.
```

```python
# Address ranges of the list
print(list_3[1:4])
print(list_3[:3])
print(list_3[12:])
```

```python
# Loop a list
for x in list_3:
	print(x)
```

```python
# Move objects from one list to another list
print(list_1)
print(list_3)
list_3.append(list_1.pop(1)) #pop mean position
print(list_1)
print(list_3)
```

```python
# Move objects from one list to a specified position in another list
print(list_1)
print(list_3)
list_3.insert(2, list_1.pop(0)) #7 has been replaced by 2
print(list_1)
print(list_3)
```

```python
for x in list_3:
	if type(x) is not int:
		print("not integer")
	else:
		print("integer")
for x in list_3:
	if isinstance(x, int):
		print("integer")
	else:
		print("not integer") 
```
